```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips["total_bill"], kde=False, bins=20)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips, kind="reg")
```

```python
sns.pairplot(tips, hue="sex", palette="coolwarm")
```

```python
sns.rugplot(tips["total_bill"])
```

```python
import numpy as np
```

```python
sns.kdeplot(tips["total_bill"])
```

```python
sns.barplot("sex", "total_bill", data=tips, estimator=np.std)
```

```python
 sns.countplot("sex", data=tips)
```

```python
sns.boxplot("day", "total_bill", data=tips)
```

```python
sns.violinplot("day", "total_bill", data=tips, hue="smoker", split="true")
```

```python
sns.stripplot("day", "total_bill", data=tips, jitter=True, hue="sex", split=True)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split = True)
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex", split = True, color="black")
```

```python
sns.factorplot(x="day", y="total_bill", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", markers=["o", "v"],
          scatter_kws={'s':100})
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, row="day", col = "time", 
           hue="sex", aspect=0.6, size=8)
```

```python
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
